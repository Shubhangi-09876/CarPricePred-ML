```python
m1 = {'Diesel' : 0, 'Petrol' : 1, 'CNG' : 2, 'LPG' : 3, 'Electric' : 4}
df['fuel'] = df['fuel'].map(m1)
```

```python
m3 = {'Individual' : 0, 'Dealer' : 1, 'Trustmark Dealer' : 3}
df['seller_type'] = df['seller_type'].map(m3)
```

```python
m4 = {'First Owner' : 0, 'Second Owner' : 1, 'Third Owner' : 2, 'Fourth & Above Owner' : 3, 'Test Drive Car' : 4}
df['owner'] = df['owner'].map(m4)
```

```python
X = df.drop(['name', 'selling_price'], axis = 1)
Y = df['selling_price']
```

```python
X
```

```python
Y
```

```python
# puting 70% data in training and 30% in testing

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.1, random_state = 2)
```

```python
lin_reg_model = LinearRegression()
```

```python
lin_reg_model.fit(X_train, Y_train)
```

```python
training_data_prediction = lin_reg_model.predict(X_train)
```

```python
error_score = metrics.r2_score(Y_train, training_data_prediction)
print("R squared Error : ", error_score)
```
